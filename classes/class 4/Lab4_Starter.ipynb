```python
%pip install matplotlib numpy

import matplotlib.pyplot as plt
import numpy as np
import time
import random
```

```python
def min1(input_list: list[int])->int:
    min = input_list[0]
    for value in input_list:
        if min > value :
            min = value
    return min
```

```python
def min2(input_list: list[int])->int:
    # performs bubble sort on the input
    n = len(input_list)
    for i in range(n):
        for j in range(1, n-i):
            if input_list[j - 1] > input_list[j]:
            # swap elements
                temp = input_list[j - 1]
                input_list[j - 1] = input_list[j]
                input_list[j] = temp
    return input_list[0]
```

### Step 1:
In a main code block, write code that creates a list of 5000 values, and initializes them to random int values (you choose the range of values). Invoke the function min1() one time on this array, timing how long it takes to execute, in seconds. Put the answer in your lab submission.

```python
def main():
    size = 5000
    list1 = []
    for i in range(size):
        list1.append(random.randint(0,1000))

    start = time.perf_counter()
    min1(list1)
    stop = time.perf_counter()

    print(stop-start)

if __name__ == "__main__":
    main()

#It took 0.0001690830000029564 second to run the program
```

```python
def main():
    size = 1000
    list1 = []
    tpc = []
    
    tets = time.perf_counter()
    for i in range(1,11):
        tpcs = time.perf_counter()
        for j in range(size*i):
            list1.append(random.randint(0,10000))
        min1(list1)
        tpce = time.perf_counter()
        tpcr = tpce-tpcs
        tpc.append(tpcr)
    tete = time.perf_counter()

    tet = tete - tets

    print("n:    time per call:     total elapsed time:")
    key = ""
    for f in range(len(tpc)):
        key = f"{key}{1000*(1+f)}     {tpc[f]}    {tet}"
        key = f"{key} + \n"

    print(key)

if __name__ == "__main__":
    main()

# n:    time per call:     total elapsed time:
# 1000     0.0008746670000618906    0.0429591250001522 
# 2000     0.0024221249998390704    0.0429591250001522 
# 3000     0.0022863750000396976    0.0429591250001522 
# 4000     0.0028412920000846498    0.0429591250001522
# 5000     0.0036970830001337163    0.0429591250001522 
# 6000     0.004496416000165482    0.0429591250001522 
# 7000     0.005261083000050348    0.0429591250001522 
# 8000     0.006106582999791499    0.0429591250001522 
# 9000     0.006900000000086948    0.0429591250001522  
# 10000     0.008067458000141414    0.0429591250001522 
```
